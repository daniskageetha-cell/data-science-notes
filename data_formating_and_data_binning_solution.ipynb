Data Binning

 why binning?

 it is process of transforming continuous numerical variables into into discrete categorical 'bins' groups

```python
df['horsepower']=df['horsepower'].astype(float)
df['horsepower'].dtypes
```

```python
import matplotlib.pyplot as plt
plt.hist(df["horsepower"])
plt.ylabel('count')
plt.xlabel('horsepower')
plt.title('horsepower bins')
```

```python
# create bins
bins = np.linspace(min(df["horsepower"]),max(df["horsepower"]),4)
bins
```

```python
#create group names.
group_names = ['Low','Medium','High']
```

```python
#perform binning
df['horsepower-binned']=pd.cut(df['horsepower'],bins, labels=group_names)
df[["horsepower","horsepower-binned"]].head(20)
```

```python
j=df["horsepower-binned"].value_counts()
j
```

```python
plt.bar(group_names,j)
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
```
